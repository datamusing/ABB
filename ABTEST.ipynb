A/B test function 
----

Lets make a litte function that would return the mean $p$ and the standard error $SE$  given conversion number and sample size. 

```python
def abFunction(conversions, sampleSize):
    p = conversions/float(sampleSize)
    q = 1. - p 
    se = np.sqrt(float(p)*float(q)/float(sampleSize))
    return p, se                            
```

Computing the mean and standard errors of control and treatment groups 
----

Our Null Hypothesis here is there is no difference between the control and treatment group. 
Our hope it to be able to reject the null hypothesis i.e. to say with enough confidence that the variation did lead to a signifant change in the conversion rate. To do this, we need to compare  the sample means of the treatment and control groups and test whether they are significantly different. To do that, we compute the 95% confidence interval of each mean (the interval within which the mean can randomly occur 95% of times the experiment is repeated).  If the two intervals are non-overlapping, then we can be sure beyond reasonable doubt that there is a real effect in the treatment group.  The confidence interval is defined as $(M \pm 1.96\times {SE})$ where  $M$ is sample mean, and $SE$ is the standard error. The asuumption here is that the sample mean is distributed as a normal variable, which is a very good approximation for a large sample such as this one. 



First,  we compute the sample mean $p$ and the confidence interval for the control group:

```python
p_c, se_c = abFunction(control.converted.sum(), control.shape[0])
print "The control group conversion rate is (%f +/- %f) percent."%(p_c*100., se_c*1.96*100.)
```

Next, we do the same for the treatment group:

```python
p_t, se_t = abFunction(treatment.converted.sum(), treatment.shape[0])
print "The treatment group conversion rate is (%f +/- %f) percent."%(p_t*100., se_t*1.96*100.)
```

Therefore, we see that the confidence interval for the conversion rate in the control is $(10.04 \pm 0.19)$. While the same for the treatment sample is  $(10.14 \pm 0.19)$. Therefore, although there seems to be an improvement in the conversion in the mean, it is not significant at all, the the intervals overlap a lot. So *we cannot conclude yet* that the variation lead to an improved conversion rate. 


Another way of seeing this is to compute the $Z$-score: 

Z-Score 
----

$Z$ score computes the distance between the means of the control and treatment group in terms of the standard errors of each sample:

$$ z = \frac{(M_1 - M_2)}{(SE_1^2 + SE_2^2)^{1/2}} $$

Ideally for $95\%$ confidence (two-tailed) we would like the have a $Z$ value of 1.96 (1.645 for one-tailed). 

The probabilty of the two means being different by chance is given by:


$ 1- \rm{Probabilty}(\lt z) $ if $\rm{Probabilty}(\lt z) \ge 0.5$

or 

$\rm{Probabilty}(\lt z)$ if $\rm{Probabilty}(\lt z) \lt 0.5$

Lets compute our Z score:



```python
Z_score = (p_t - p_c)/sqrt(se_c**2+se_t**2)
Z_score
```

Therefore, we see that Z-score is only 0.7, which means that there is a $30\%$ chance that the means are different just due to random fluctuations. Hence, the difference is not significant at all. 

How large a sample do we need?
-----

The next obvious question is how large a test sample we would need to make a significant detection. An easy way to estimate this is to say that both groups have roughly the same $SE = SE_1 \simeq SE_2$, and then demand for a Z-score of 1.96. This gives, from the definition of z:


$$ 1.96 = \frac{(M_1 - M_2)}{(2 ~ SE^2)^{1/2}} $$ 

Remembering that $SE^2 = p~(1-p)/N$, we have:


$$ N = 2\times 1.96^2  p~(1-p) /(M_1 -M_2)^2$$ per branch (either control or treatment).


Let's compute  this number:

```python
diff = 0.001
se0  = sqrt(0.10*(1.-0.10))
N = (np.sqrt(2)*1.96*se0/diff)**2
print N
```

Therefore, in each branch we need 691488 samples, meaning that in total for a 50/50 split we need twice this value i.e. 1382976 samples. Currently, we only have about 191000 samples which is clearly 6 times smaller than what we need.
