A/B test function 
=======

Lets make a litte function that would return the mean p-value and the standard error  given conversion number and sample size. 

```python
def abFunction(conversions, sampleSize):
    p = conversions/float(sampleSize)
    q = 1. - p 
    se = np.sqrt(float(p)*float(q)/float(sampleSize))
    return p, se                            
```

The std dev per sample is $\sigma = \sqrt{(p (1-p)}$. For a sample of size $N$, I have the standard error of either control or treatment as  $\sigma_N= \sigma/\sqrt{N}$. 
The se of the difference is $\sqrt 2 \sigma_N$
For 95% significance, I want the difference in the two means to be at least $\Delta = \sqrt{2}\times 1.96 \sigma_N$ which implies 

$N = 2\sigma^2 1.96^2 /\Delta^2 $ samples per branch!
